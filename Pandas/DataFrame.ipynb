```python
import numpy as np
import pandas as pd

# Dataframe is a 2D table format data. 

# A) Creating a DataFrame from a List

data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]

col = ['Name','Age','City','Salary'] # we are providing labels here for the data frame

df_from_list = pd.DataFrame(data_list,columns=col)
```

```python
df_from_list
```

```python
# Creating a DataFrame using Dictionary

data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}

df_from_dict = pd.DataFrame(data) # Here also The labels are given as per the DICT KEYS
```

```python
df_from_dict
```

## Selection and Indexing


```python
df_from_dict
```

```python
# SELECTING A SINGLE ROW
df_from_dict['Name'] # PROVIDE COL NAME 
```

```python
# SELECTING MULTIPLE ROW
df_from_dict[['Age','Salary']] #give col name in a nested list format
```

## Creating a new column

```python
df_from_dict
```

```python
#Adding a new col as Country

# We need to provide data for each row else value error is occured
#df_from_dict["Country"] = ["India", "Korea"]  #ValueError: Length of values (2) does not match length of index (4)

df_from_dict['Country'] = ['India','Korea','Japan','Ireland']
df_from_dict
```

## Removing a Column from DataFrame

```python
df_from_dict
```

```python
# We remove the country col
# we need to provide AXIS for finding the col , by Default axis=0 
df_from_dict.drop('Country',axis=1)
```

```python
# IMPORTANT : but if we again print the dataframe , we see country is still present into it. 
# Because it only show the snapshot or a copy of ACTUAL DATAFRAME and it remain in the storage
df_from_dict
```

```python
# Now if we want to completely remove the country column , we use INPLACE = True in .drop()
df_from_dict.drop('Country', axis=1, inplace=True)
```

```python
df_from_dict
```

```python
# to remove multiple rows 
df_from_dict.drop(['Salary','City'], axis=1)
```
