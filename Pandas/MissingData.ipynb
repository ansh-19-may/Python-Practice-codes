## Finding Missing Values



### NOTE: YOU NEED TO GIVE inplace=True FOR ACTUAL MODIFICATION OF TABLE. WITHOUT IT , MODIFICATION ONLY DONE ON COPY OF REAL DATA


```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Now if we want a bool table to know about nan data , we use .isna()

df.isna()
```

```python
# now if we want to know how many NaN values are in each col . we use .isna().sum()
df.isna().sum()
```

```python
# if we want to know is there any column with no NaN values . we use .isna().any()

df.isna().any()
```

## Removing Missing Values 
### Removal of missing values happens using rows of dataframe, which means row-wise removal happens.

```python
df
```

```python
# dropna() is used to remove

df.dropna()
```

```python
# dropna(thresh=3) here thresh har ek row me jake check karta hai ki ATLEAST 3 NON NULL values toh present honi chahiye. if present hai toh vo rows ko conserve karo
# aur agar 3 non null values se kam ho toh delete kardo.

df.dropna(thresh=3)
```

## Filling Missing Values

```python
df
```

```python
df.fillna(0)
```

```python
fill_values = {'A':10,'B':20,'C':30,'D':40}

df.fillna(value=fill_values)
```

```python
df.fillna(df.mean())
```
