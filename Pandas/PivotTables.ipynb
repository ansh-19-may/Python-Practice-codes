## Pivot Tables and Cross Tab


### Pivot table gives AGGREGATED VALUES such as mean, median etc and Cross tab gives COUNT VALUES as result.

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values="Sales",index="Region",columns="Product", aggfunc="mean")

# we provide dataframe.
# in values= we add column name on which the any aggregated value in .done, for ex Sales, here Mean of Sales in each REGION is done, By default Mean is taken. or use aggfunc= 
# index= here we add column Name as the index for getting index name.
# columns= here we apply diffrent kinds of Product as separate cols on which sales of each product in each region is displayed. 
```

```python
pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
```

## Cross Tabs


```python
pd.crosstab(df['Region'],df['Product'])
```
